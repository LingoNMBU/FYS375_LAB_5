```python
path2 = 'LAB5_rawdata_2018(2).csv'
df2= pd.read_csv(path2, sep=';', decimal=",")
df2['Time'] = pd.to_datetime(df2['Time'], infer_datetime_format=True)
df_day = df2[df2['Time'].dt.strftime('%Y-%m-%d') == "2018-03-26"]
df_day
```

```python
df_day.index = df_day['Time']

df_day = df_day.between_time('8:00', '19:00')

df_day = df_day.drop('Time', axis=1)
```

```python
area = 1.665*0.991*9
R_wire = 0.086
```

```python
#df2['Input Power'] = df2['Plane of Array Irradiance (W/m2)']*area

df_day['Input Power']  = df_day.apply(lambda row: row['Plane of Array Irradiance (W/m2)']*area , axis=1)
df_day['wire_loss']  = df_day.apply(lambda row: (row['Inverter PV Current (ADC)']**(2))*R_wire , axis=1)
df_day['eta_wire']  = df_day.apply(lambda row: 1 -row['wire_loss']/row['Inverter PV Power (WDC)'] , axis=1)
```

```python
df_day.columns
```

```python
df_day['eta_sun_mod'] = (df_day['Inverter PV Power (WDC)']+df_day['wire_loss'])/df_day['Input Power']
df_day['eta_inv'] = df_day['Output Power (W)']/df_day['Inverter PV Power (WDC)']
df_day['eta_system'] = df_day['Output Power (W)']/df_day['Input Power']
```

```python
fig, ax = plt.subplots(figsize = (4, 3))
ax.set_ylim(0.05,0.16)
ax.set_ylabel('Efficiency')
ax.set_xlabel('Output Power [W]')
plt.tight_layout()

plt.scatter(df_day['Output Power (W)'], df_day['eta_sun_mod'], s=15)
plt.savefig('Efficiency_sun_to_mod.pdf')
```
